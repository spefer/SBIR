```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from keras import models
from keras import layers
from tensorflow.keras import optimizers
from sklearn.metrics import mean_squared_error, r2_score
import matplotlib.pyplot as plt
```

```python
from tensorflow import random
random.set_seed(0)

data = pd.read_csv("sbir_lights_lec1_1.csv")
```

```python
df = pd.DataFrame(data)
df.head()
```

```python
X = df.iloc[:,1:2].values
y=df.iloc[:,4:5].values
```

```python
network = models.Sequential()
network.add(layers.Dense(32, activation='relu', input_shape=(1,)))
network.add(layers.Dense(32, activation='relu'))
network.add(layers.Dense(32, activation = 'relu'))
network.add(layers.Dense(32,activation = 'relu'))
network.add(layers.Dense(1))

network.compile(optimizer='adam',
                loss='mse',
                metrics=['mae'])
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

Standardize and train the model

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)

history = network.fit(X_train, y_train,
                    validation_data=(X_test, y_test),
                    epochs=250,
                    batch_size=5)
```

```python
history_dict = history.history
loss_values = history_dict['loss']
val_loss_values = history_dict['val_loss']
accuracy = history_dict['mae']
val_accuracy = history_dict['val_mae']
 
epochs = range(1, len(loss_values) + 1)
```

**Plot The Training & Validation**

```python
fig, ax = plt.subplots(1, 2, figsize=(14, 6))

import warnings
warnings.filterwarnings("ignore")

# Plot the model MAE vs Epochs
ax[0].plot(epochs, accuracy, 'bo', color="blue", label='Training MAE')
ax[0].plot(epochs, val_accuracy, 'b', color="orange", label='Validation MAE')
ax[0].set_title('Training & Validation MAE', fontsize=16)
ax[0].set_xlabel('Epochs', fontsize=16)
ax[0].set_ylabel('MAE', fontsize=16)
ax[0].legend()

# Plot the loss vs Epochs
ax[1].plot(epochs, loss_values, 'bo', color="blue", label='Training loss')
ax[1].plot(epochs, val_loss_values, 'b', color="orange", label='Validation loss')
ax[1].set_title('Training & Validation Loss', fontsize=16)
ax[1].set_xlabel('Epochs', fontsize=16)
ax[1].set_ylabel('Loss', fontsize=16)
ax[1].legend()
```

```python
y_test_pred = network.predict(X_test)
print("R2 Score : ", r2_score(y_test, y_test_pred)*100)
print('Mean Squared Error (MSE):', mean_squared_error(y_test, y_test_pred))
```

```python
plt.figure(figsize=(10,10))
plt.scatter(y_test, y_test_pred, c='crimson')

p1 = max(max(y_test_pred), max(y_test))
p2 = min(min(y_test_pred), min(y_test))
plt.plot([p1, p2], [p1, p2], 'b-')
plt.xlabel('True Values', fontsize=15)
plt.ylabel('Predictions', fontsize=15)
plt.axis('equal')
plt.show()
```

```python
compare = pd.DataFrame({"Actual": y_test.flatten(), "predicted": y_test_pred.flatten()})
print(compare)
```

```python
Adj_r2 = 1 - (1-r2_score(y_test, y_test_pred)) * (len(y)-1)/(len(y)-X.shape[1]-1)
print("Adjusted R^2 :", Adj_r2)
```

**Residual plot**

```python
residuals=y_test-y_test_pred
plt.scatter(y_test_pred,residuals)
plt.xlabel('Predicted values')
plt.ylabel('Residuals')
plt.axhline(y=0, color='r', linestyle='-')
plt.show()
```
